# Estudo do tempo 

```python
import pandas as pd
tempo_dia = pd.read_csv('../data/times/times_1_day.csv')
tempo_hora = pd.read_csv('../data/times/times_1_hour.csv')
tempo_minuto = pd.read_csv('../data/times/times_10_minutes.csv')

tempo_minuto['time'] = pd.to_datetime(tempo_minuto['time'], utc = True)
tempo_hora['time'] = pd.to_datetime(tempo_hora['time'], utc = True)
tempo_dia['time'] = pd.to_datetime(tempo_dia['time'], utc = True)
```

```python
tempo_minuto[tempo_minuto['id_time'] > 14610]
```

```python
lista_minutos = list(tempo_minuto['time'].dt.minute)
lista_IDS_minutos = list(tempo_minuto['id_time'])
lista_horas = list(tempo_hora['time'].dt.hour)
lista_IDS_horas = list(tempo_hora['id_time'])

lista_dias = list(tempo_dia['time'].dt.day)
lista_IDS_dias = list(tempo_dia['id_time'])

diff_d = {}
for i in range( len(lista_dias) -1):
    diff_d[lista_IDS_dias[i+1]] = lista_dias[i+1] - lista_dias[i]

diff_m = {}
for i in range( len(lista_minutos) -1):
    diff_m[lista_IDS_minutos[i+1]] = lista_minutos[i+1] - lista_minutos[i]

diff_h = {}
for i in range( len(lista_horas) -1):
    diff_h[lista_IDS_horas[i+1]] = lista_horas[i+1] - lista_horas[i]

print("minutos")
for m in list(diff_m.keys()):
    if not((int(m)%6 == 0 and int(diff_m[m]) == -50) or int(diff_m[m]) == 10):
        print(f"{m}  :  {diff_m[m]}")

print("hora")
for h in list(diff_h.keys()):
    if not((int(h)%24 == 0 and int(diff_h[h]) == -23) or int(diff_h[h]) == 1):
        print(f"{h}  :  {diff_h[h]}")

print("dia")
for d in list(diff_d.keys()):
    if not((int(d)%31 == 0 and int(diff_d[d]) == -30 ) or int(diff_d[d]) == 1):
        print(f"{d}  :  {diff_d[d]}")
```
