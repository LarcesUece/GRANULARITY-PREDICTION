O RMSE é influenciado apenas pelo modelo, ou também pela granularidade temporal (DIM), e existe interação entre eles?

```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf

df = pd.read_csv("../../MODELOS/modelos_treinados.csv")

# Garantir que são categóricas
df["MODELO"] = df["MODELO"].astype("category")
df["DIM"] = df["DIM"].astype("category")

# RMSE ~ MODELO + DIM + MODELO:DIM
model = smf.ols("RMSE ~ C(MODELO) + C(DIM) + C(MODELO):C(DIM)", data=df).fit()

print(model.summary())
```
